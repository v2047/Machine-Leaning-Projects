**About the data:**

Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value **1 in case of fraud 0 otherwise**.


Time : time elasped in second

1 - fraudulent transactions

0 - valid transactions

# Under Sampling


Build a sample dataset containing similar distribution of valid transaction and fraudulent transactions


No of fraudulent transactions  = 492

```python
legit_sample = legit.sample(n=492) # here we take the random 492 data (sample) from vaild transaction
```

Concatenating two DataFrames

axis =0 : values add by rows

axis =1 : values add by columns

```python
new_credit_data = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_credit_data.head()
```

```python
new_credit_data.tail()
```

```python
new_credit_data["Class"].value_counts()
```

Now we have the uniform data which is good for buliding the good ML Model.


```python
new_credit_data.groupby("Class").mean()
```

Based on the above observatio ,we found that the mean values remain nearly unchanged(for Balanced and unbalaned data, mean is almost same)that means dataset nature is not changed.This step is important to check whether we have good sample or bad sample.
In case of bad sample , we should have received change in means.

Splitting the data into Features & Targets



```python
# features
X = new_credit_data.drop(columns = 'Class', axis =1)
print(X.head())
```

```python
# Traget
Y = new_credit_data['Class']
```

```python
print(Y)
```

Spliting the data into Training and Test


```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

Logistics Regression Model as Y (target/dependent variable) is binary.

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

**Model Evaluation**

Accuracy Score


```python
# Accuracy on Training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Tarining Data', training_data_accuracy)
```

If Accuracy Score is 0.75 or higher than this i.e. prediction is good.
The accuracy is calculated by comparing the predicted labels (y_pred) with the true labels (y_test).

```python
# Accuracy on test data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)



```

```python
print('Accuracy test Data', testing_data_accuracy)
```

Accuracy is good for test and training data that means Model is good.

If Accuracy test and traning has differrnt value i.e. model is underfitting or overfitting.
